```python
import pandas as pd
import numpy as np
data= pd.read_csv('wnbastats.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#plot the teams' turnovers vs their assists
sns.lmplot( data=data, x="turnover", y="assist", hue="team", col="team")
```

```python
#plot the teams' defensive rebounds along with their total rebounds
f, ax= plt.subplots(figsize=(7,5))
sns.set_color_codes("pastel")
sns.barplot(x="totrb", y="team", data=data, label="Total Rebounds", color="b") #plot the total rebounds
sns.set_color_codes("muted") 
sns.barplot(x="defrb", y="team", data=data, label="Defensive Rebounds", color="r") #plot the defensive boards
ax.set(xlim=(0,7), xlabel="Total Rebounds by teams")
```

```python
#plot the teams' offensive rebounds and total rebounds 
f, ax= plt.subplots(figsize=(7,5))
sns.barplot(x="totrb", y="team", data=data, label="Total Rebounds", color="b") #plot the total rebounds
sns.set_color_codes("muted")
sns.barplot(x="offrb", y="team", data=data, label="Offensive Rebounds", color="r" ) #plot the offensive rebounds 
ax.set(xlim=(0,7),xlabel="Total Rebounds by Teams")
```

```python
#plot the team's attempted 3 point attempts vs their made 3 pointers 
f, ax= plt.subplots(figsize=(10,10))
sns.barplot(x="att3", y="team", data=data, label="Attempted 3 pointers", color="y") 
sns.set_color_codes("muted")
sns.barplot(x="made3", y="team", data=data, label="Made 3 pointers", color="g" )  
ax.set(xlim=(0,6),xlabel="Total 3 Pointers")
```

```python
#plot the team's field goal attempts vs their made field goals 
sns.set_theme(style="darkgrid")
sns.displot( data=data, x="fgatt", y="fgmade", col="team", binwidth=5, height=5)
```

```python
#plot the average point of the teams influenced by their wins/losses
sns.set_theme(style="darkgrid")
sns.catplot( data=data, kind="bar", x="team", y="team_pts", hue="win", errorbar="sd", palette="dark:m_r", alpha=.6, height=6)
```

```python
#plotting player 3 point attempts vs 3 pointers made 
f, ax= plt.subplots(figsize=(20,50))
sns.barplot(x="att3", y="Player", data=data, label="Attempted 3 pointers", color="b") 
sns.barplot(x="made3", y="Player", data=data, label="Made 3 pointers", color="r" )  
ax.set(xlim=(0,15),xlabel="Total 3 Pointers")
```

```python
#teams' minutes vs their points for the season 
sns.lmplot(data=data, x="minutes", y="team_pts", hue="team", height=25)
```

```python
#plotting the player id against the player steals 
sns.set_theme(style="darkgrid")
f, ax= plt.subplots(figsize=(10,10))
sns.barplot(data=data, x="team", y="steal", hue="team")
```
